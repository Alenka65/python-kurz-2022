```python
import pandas
temperature = pandas.read_csv('temperature.csv')
temperature.info()
```

```python
temperature.head()
```

Dotaz na měření, která byla provedena v Praze

```python
temperature[temperature['City'] == "Prague"]
```

Dotaz na měření, ve kterých je teplota (sloupec AvgTemperature) vyšší než 80 stupňů.

```python
tabulka = temperature[temperature['AvgTemperature'] > 80]
tabulka.reset_index(drop=True)
```

Dotaz na měření, ve kterých je teplota vyšší než 60 stupňů a současně bylo měření provedeno v regionu (sloupec Region) Evropa (Europe).

```python
tabulka1 = temperature[(temperature['AvgTemperature'] > 60) & (temperature['Region'] == 'Europe')]
tabulka1.reset_index(drop=True)
```

Dotaz na extrémní hodnoty, tj. měření, ve kterých je teplota vyšší než 80 stupňů nebo menší než -20 stupňů.

```python
tabulka3 = temperature[(temperature['AvgTemperature'] > 80) | (temperature['AvgTemperature'] < -20)]
tabulka3.reset_index(drop=True)
```
